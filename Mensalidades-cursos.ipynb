```
# Importando Blibliotecas
import pandas as pd
# Magica do matpltolib
%matplotlib inline

df = pd.read_csv('https://raw.githubusercontent.com/rodolfoghi/curso-labhacker-analise-dados-com-python/master/cursos-prouni.csv')
df.head()
```

```
# Verificar o tamanho do DataFrame
# Formato (linhas, colunas)

df.shape
```

```
# Tipos de dados das colunas
df.dtypes
```

```
# Analise  Matematica das colunas numericas

df.describe()
```

```
# Descreverndo os dados
df.info()
```

```
# Quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```
# Qual curso com a mensalidade mais alta?
#subgrupo para pegar a coluna
df[df['mensalidade'] == df['mensalidade'].max()]
```

```
# Qual o curso com menor mensalidade, mais barato do Brasil?
df[df['mensalidade'] == df['mensalidade'].min()]
```

```
# Qual a mensalidade mais cara por UF
df.groupby('uf_busca')['mensalidade'].max()
```

```
# Qual a mensalidade mais barata por UF?
df.groupby('uf_busca')['mensalidade'].min()
```

```
# Mostrar todos os dados cursos mais caros de cada Uf?
# loc, comando do data frame que vai localizar uma linha
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]
```

```
# EMnsalidades entre 500 e 1000 reais

df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000 )]
```

```
# Quais as 5 maiores mensalidade?
df.nlargest(5, 'mensalidade')
```

```
# Qual são as 5 menores mensalidades?
df.nsmallest(5, 'mensalidade')
```
